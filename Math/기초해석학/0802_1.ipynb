# Scatter

```python
np.random.seed(19680801)

num = 100
x = np.random.rand(num) # 0 ~ 1 사이 실수로 반환
y = np.random.rand(num)

plt.scatter(x, y)
plt.show()
```

```python
np.random.seed(842344)

num = 100
x = np.random.rand(num)
y = np.random.rand(num)
area = (30 * np.random.rand(num)) ** 2

plt.scatter(x, y, s=area) # s=area => 점의 원 크기를 변환하는 키워드
plt.show()
```

```python
np.random.seed(842344)

num = 100
x = np.random.rand(num)
y = np.random.rand(num)
area = (30 * np.random.rand(num)) ** 2

plt.scatter(x, y, s=area, alpha=0.5) # alpha=0.5 => 투명도 0.5
plt.show()
```

```python
np.random.seed(842344)

num = 100
x = np.random.rand(num) # 100개의 0 ~ 1사이 실수를 배열의 형태로 반환
y = np.random.rand(num)
colors = np.random.rand(num)
area = (30 * np.random.rand(num)) ** 2

plt.scatter(x, y, s=area, c=colors, alpha=0.5) # c=colors => 색상 변경
plt.show()
```

# 단일 변수 함수

```python
x = np.arange(0., 5., 0.2) # 정의역 0.0 ~ 5.0, 0.2 간격으로
y = x**2 # x의 제곱

plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0., 5., 0.2)
y1 = x
y2 = 0.25 * x ** 2
y3 = np.cos(x)

plt.plot(x, y1, '-o') # '-o' => solid 줄모양에 동그라미 모양
plt.plot(x, y2, '--s') # '--s' => dashed 줄모양에 네모 모양
plt.plot(x, y3, ':x') # ':x' => dotted 줄모양에 X 모양
plt.legend(["y = x", "y = 0.25 * x ** 2", 'y = cos(x)']) # 좌상단의 설명문 달아주는거
plt.show()
```

```python
x = np.arange(0., 5., 0.2)
y1 = x
y2 = 0.25 * x ** 2
y3 = np.cos(x)

plt.plot(x, y1, '-o') 
plt.plot(x, y2, '--s') 
plt.plot(x, y3, ':x')
plt.title('plot tutorial') # 제목 달아주는거
plt.grid() # plot에 격자 달아주는거
plt.xlabel("X") # X축에 적어줄 설명
plt.ylabel("Y") # Y축에 적어줄 설명
plt.legend(["$y = x$", r"$y = \frac{1}{4}x^2$", 'y = cos(x)']) # "$x = y$" 글자 기울게하기 / r"$y = \frac{1}{4}x^2$" 4분의 1, x제곱 표현하는 방법
plt.show()
```

```python
x = np.arange(0., 5., 0.2)
y1 = x
y2 = 0.25 * x ** 2
y3 = np.cos(x)

plt.subplot(3, 1, 1)
plt.title("subplot tutorial")
plt.grid()
plt.plot(x, y1, '-ok') # '-ok' 줄모양, 줄에 들어갈 모양, 줄색상 k => black

plt.subplot(3, 1, 2)
plt.plot(x, y2, '--sr') # r => red

plt.subplot(3, 1, 3)
plt.plot(x, y3, ':xg') # g => green
plt.show()
```

# 다변수 함수

```python
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)

R = np.sqrt(X**2 + Y**2)
Z = np.sin(R) # sin(루트 X^2 + Y^2)

fig = plt.figure(figsize=(5, 5)) # plt.show할때 크기 정해주는거
plt.contour(X, Y ,Z, cmap='coolwarm') # plt.contour(X, Y ,Z, cmap='coolwarm') 선 색상 정의해주는거
plt.grid()
plt.axis('equal')
plt.show()
```

```python
fig = plt.figure(figsize=(5, 5)) 
plt.contourf(X, Y ,Z, cmap='coolwarm') # plt.contourf(X, Y ,Z, cmap='coolwarm') 선과 선 사이 색상 채워주는거
plt.grid()
plt.axis('equal')
plt.show()
```
